```python
import pandas as pd
from sklearn.model_selection import train_test_split, cross_val_score, StratifiedKFold
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.svm import SVC
from imblearn.over_sampling import RandomOverSampler
from sklearn.metrics import classification_report, accuracy_score

# Load the original dataset
data = pd.read_csv('data_public.csv')

# Separate features and target
X = data.drop(columns=['Class'])  # Assuming 'Class' is the target column
y = data['Class']

# Split data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, stratify=y, random_state=42)

# Apply Random Oversampling to the training data
ros = RandomOverSampler(random_state=42)
X_train_resampled, y_train_resampled = ros.fit_resample(X_train, y_train)

# Define models and pipelines
models = {
    'Logistic Regression': Pipeline([
        ('scaler', StandardScaler()),
        ('classifier', LogisticRegression(class_weight='balanced', max_iter=300, random_state=42))
    ]),
    'Decision Tree': Pipeline([
        ('classifier', DecisionTreeClassifier(class_weight='balanced', random_state=42))
    ]),
    'Gradient Boosting': Pipeline([
        ('classifier', GradientBoostingClassifier(random_state=42))
    ]),
    'MLP': Pipeline([
        ('scaler', StandardScaler()),
        ('classifier', MLPClassifier(hidden_layer_sizes=(128, 64), activation='relu', solver='adam',
                                      max_iter=300, random_state=42))
    ]),
    'SVM': Pipeline([
        ('scaler', StandardScaler()),
        ('classifier', SVC(kernel='rbf', class_weight='balanced', random_state=42))
    ])
}

# Define K-Fold Cross-Validation
kfold = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# Train, cross-validate, and test each model
results = {}
for name, pipeline in models.items():
    print(f"\nTraining and Cross-Validating {name}...")

    # Perform K-Fold Cross-Validation
    cv_scores = cross_val_score(pipeline, X_train_resampled, y_train_resampled, cv=kfold, scoring='accuracy')
    print(f"{name} Cross-Validation Scores: {cv_scores}")
    print(f"{name} Mean Cross-Validation Accuracy: {cv_scores.mean():.4f}")

    # Train on full training set
    pipeline.fit(X_train_resampled, y_train_resampled)

    # Test on the test set
    y_test_pred = pipeline.predict(X_test)
    test_acc = accuracy_score(y_test, y_test_pred)
    print(f"{name} Test Accuracy: {test_acc:.4f}")
    print(f"\nClassification Report for {name} (Test Set):")
    print(classification_report(y_test, y_test_pred))

    # Save results
    results[name] = {
        'Cross-Validation Accuracy': cv_scores.mean(),
        'Test Accuracy': test_acc
    }

# Display overall results
print("\nOverall Results:")
for model, metrics in results.items():
    print(f"{model}: Cross-Validation Accuracy = {metrics['Cross-Validation Accuracy']:.4f}, "
          f"Test Accuracy = {metrics['Test Accuracy']:.4f}")
```
